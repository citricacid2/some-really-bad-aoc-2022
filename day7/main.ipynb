```python
with open("input.txt") as f:
    f.readline()
    data = f.read().splitlines()
data
```

```python
def get_folder(folder_path):
    current_dir = ""
    for index, line in enumerate(data):
        if line.startswith("$"):
            # print(line)
            line = line[2:]
            if line.startswith("cd"):
                line = line[3:]
                if line == "..":
                    current_dir = current_dir[:current_dir.rfind("/")]
                else:
                    current_dir += "/" + line  
                # print(current_dir)
            elif line.startswith("ls"):
                folder = []
                checker_index = index + 1
                while checker_index < len(data) and not data[checker_index].startswith("$"):
                    folder.append(data[checker_index])
                    checker_index += 1
                # print(folder)
                # print(folder_size(folder))
                if folder_path == current_dir:
                    print(f"folder_path: {folder_path}")
                    return folder
                elif folder_path == "/":
                    return folder

current_dir = ""
sum_of_sizes = 0
loops = 0

def folder_size(folder):
    global current_dir
    global sum_of_sizes
    global loops
    size = 0
    print(f"current directory: {current_dir}")
    print(f"files in folder: {folder}")
    for file in folder:
        if file[0].isdigit():
            size += int(file.split(" ")[0])
        else:
            current_dir += "/" + file[4:]
            size += folder_size(get_folder(current_dir))
            loops += 1
            current_dir = current_dir[:current_dir.rfind("/")]
    if size <= 100000:
        sum_of_sizes += size

    return size

folder_size(get_folder("/"))
print(sum_of_sizes)
print(loops)

        




            
    
```

```python
root_size = folder_size(get_folder("/"))
full_disk = 70000000
target_free_space = 30000000
print(root_size)
current_free_space = full_disk - root_size
print(current_free_space)

current_dir = ""
current_smallest = full_disk
def folder_size2(folder):

    global current_dir
    global current_smallest
    size = 0
    print(f"current directory: {current_dir}")
    print(f"files in folder: {folder}")
    for file in folder:
        if file[0].isdigit():
            size += int(file.split(" ")[0])
        else:
            current_dir += "/" + file[4:]
            size += folder_size2(get_folder(current_dir))
            current_dir = current_dir[:current_dir.rfind("/")]
    if size < current_smallest and (size + current_free_space) >= target_free_space:
        current_smallest = size
    return size

folder_size2(get_folder("/"))
print(current_smallest)
```
