```python
with open("crates.txt") as f:
    data = f.read().splitlines()
amount_of_crates = 9
crates = [[] for i in range(amount_of_crates)]
for line in data[:-1]:
    for crate in range(amount_of_crates):
        crates[crate].append(line[1 + crate * 4])
crates
# remove blanks and reverse
for crate in range(amount_of_crates):
    crates[crate] = [x for x in crates[crate] if x != " "]
    crates[crate].reverse()



crates
```

```python
with open("input.txt") as f:
    input = f.read().splitlines()
input
```

```python
for line in input:
    print(line)
    from_string_start = line.find(" from")
    
    amount_to_move = int(line[5:from_string_start])
    print(amount_to_move)

    to_string_start = line.find(" to")
    from_crate = int(line[from_string_start + 5:to_string_start])
    print(from_crate)

    to_crate = int(line[to_string_start + 3:])
    print(to_crate)
    crates_to_move = crates[from_crate - 1][-amount_to_move:]
    crates[from_crate -1] = crates[from_crate - 1][:-amount_to_move]
    crates[to_crate -1] = crates[to_crate -1] + crates_to_move


# print the top of each crate
for crate in crates:
    print(crate[-1], end="")
```
