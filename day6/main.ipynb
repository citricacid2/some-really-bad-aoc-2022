```python
with open("input.txt") as f:
    data = f.read()
data
```

```python
amount = 14
last = []
count = 0
for i in data: 
    count += 1
    if len(last) == amount:
        if len(last) == len(set(last)):
            break
        last.pop(0)
    last.append(i)
    print(f"the last 4 are {last}, and {count} characters have been read")
```
