```python
with open("input.txt") as f:
    data = f.read().splitlines()

data
```

```python
total_overlaps = 0
for i in data:
    elf1, elf2 = i.split(",")
    elf1start, elf1end = elf1.split("-")
    elf2start, elf2end = elf2.split("-")
    elf1start = int(elf1start)
    elf1end = int(elf1end)
    elf2start = int(elf2start)
    elf2end = int(elf2end)
    print(elf1start, elf1end, elf2start, elf2end)
    if (elf1start <= elf2start and elf2end <= elf1end) or (elf2start <= elf1start and elf1end <= elf2end):
        print("overlap")
        total_overlaps += 1

total_overlaps
```
