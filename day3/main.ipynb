```python
with open("input.txt") as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines]
    # split lines into groups of 3
```

```python
from string import ascii_lowercase
from string import ascii_uppercase
```

```python
total_priority = 0


for line in lines:
    length = len(line)
    first_half = line[:length//2]
    second_half = line[length//2:]
    for i in first_half:
        if i in second_half:
            print("shared letter: ", i)
            if i in ascii_lowercase:
                total_priority += ascii_lowercase.index(i) + 1
                print("priotity: ", ascii_lowercase.index(i) + 1)
            else:
                total_priority += ascii_uppercase.index(i) + 1 + 26
                print("priotity: ", ascii_uppercase.index(i) + 1 + 26)
            break
total_priority
```
