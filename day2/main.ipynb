```python
data = data.split("\n")
```

```python
total_score = 0
for i in data[:-1]:
    opponent = i[0]
    if opponent == "A":
        opponent_move = "Rock"
    elif opponent == "B":
        opponent_move = "Paper"
    elif opponent == "C":
        opponent_move = "Scissors"
    
    print(opponent_move)

    us = i[2]
    if us == "X":
        our_move = "Rock"
    elif us == "Y":
        our_move = "Paper"
    elif us == "Z":
        our_move = "Scissors"
    
    print(our_move)

    

    if opponent_move == our_move:
        score = 3
    elif opponent_move == "Rock" and our_move == "Paper":
        score = 6
    elif opponent_move == "Rock" and our_move == "Scissors":
        score = 0
    elif opponent_move == "Paper" and our_move == "Rock":
        score = 0
    elif opponent_move == "Paper" and our_move == "Scissors":
        score = 6
    elif opponent_move == "Scissors" and our_move == "Rock":
        score = 6
    elif opponent_move == "Scissors" and our_move == "Paper":
        score = 0
    
    if our_move == "Rock":
        score += 1
    elif our_move == "Paper":
        score += 2
    elif our_move == "Scissors":
        score += 3

    print(score)
    total_score += score

print(total_score)
```
